## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()


# Ignoring specific warnings
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings("ignore", category=FutureWarning)


# For checking available datasets
import os
print(os.listdir('../Data'))
```

## Import datasets

```python
titanic_train_df = pd.read_csv('../Data/titanic-survival-prediction-data/train.csv')
```

### Using Heatmap for checking NaN values in columns

```python
fig, ax = plt.subplots(figsize=(9,5))
sns.heatmap(titanic_train_df.isnull(), cbar=False, cmap="YlGnBu_r")
plt.show()
```

### Seaborn Countplot

```python
cols = ['Survived', 'Sex', 'Pclass', 'SibSp', 'Parch', 'Embarked']

nr_rows = 2
nr_cols = 3

fig, axs = plt.subplots(nr_rows, nr_cols, figsize=(nr_cols*3.5,nr_rows*3))

for r in range(0,nr_rows):
    for c in range(0,nr_cols):  
        
        i = r*nr_cols+c       
        ax = axs[r][c]
        sns.countplot(titanic_train_df[cols[i]], hue=titanic_train_df["Survived"], ax=ax)
        ax.set_title(cols[i], fontsize=14, fontweight='bold')
        ax.legend(title="survived", loc='upper center') 
        
plt.tight_layout()
```
